```python
import csv
import os
from datetime import datetime
from googleapiclient.discovery import build

# 1. Configuração 
api_key = "Insira_sua_key_aqui"
youtube_service = build('youtube', 'v3', developerKey=api_key)

# 2. 5 IDs de vídeos

lista_videos = [
    "-sk57kOoHHo",  # Forte Mão - Projeto Sola
    "PFku1zDYIns",  # O Nosso General é Cristo - PAX
    "sg3ycPXj6vA",  # Festa do Universo - Israel S.
    "tFn8561bnYY",  # Espelhos Mágicos - Oficina G3
    "Sr24c2KcbWM",  # Os Anjos te Louvam - Grupo Versos
]

# 3. Criar e abrir o arquivo Excel (CSV) para escrever 
print("Iniciando a Coleta de Dados...")

# --- PROCURANDO NAS PASTAS ---
base_usuario = os.path.expanduser("~") # Pega C:\Users\...

# Tentativa 1: Desktop normal
pasta_destino = os.path.join(base_usuario, "Desktop")

# Tentativa 2: Desktop no OneDrive (Muito provável)
if not os.path.exists(pasta_destino):
    pasta_destino = os.path.join(base_usuario, "OneDrive", "Desktop")

# Tentativa 3: Se nada der certo, salva na pasta do usuário direto
if not os.path.exists(pasta_destino):
    pasta_destino = base_usuario

# Monta o nome final
caminho_final = os.path.join(pasta_destino, "relatorio_youtube.csv")
print(f"Arquivo alvo: {caminho_final}")

# Agora sim, O Histórico
# Verifica se o arquivo já existe ANTES de abrir
arquivo_existe = os.path.isfile(caminho_final)

# Abre o arquivo em modo de adição (append) para preservar o histórico 
with open(caminho_final, mode='a', newline='', encoding='utf-8') as arquivo_csv:
    escritor = csv.writer(arquivo_csv, delimiter=';')

    # 4. Só escreve o cabeçalho se o arquivo for NOVO (não existir)
    if not arquivo_existe:
        escritor.writerow(['Data_Coleta', 'ID_Video', 'Titulo', 'Visualizacoes', 'Likes', 'Comentarios'])
        print("Arquivo novo criado! Cabeçalho inserido.")
    else:
        print("Arquivo já existe! Adicionando novos dados ao final...")

    
    # 5. O Loop de Coleta 
    for id_atual in lista_videos:
        try:
            # Pedido ao YouTube
            request = youtube_service.videos().list(
                part="snippet,statistics",
                id=id_atual
            )
            response = request.execute()

            # Verificar se o vídeo existe
            if response['items']:
                dados = response['items'][0]

            # Extraindo as informações
                titulo = dados['snippet']['title']
                views = dados['statistics'].get('viewCount', 0) # .get evita erro se não tiver views
                likes = dados['statistics'].get('likeCount', 0) 
                comments = dados['statistics'].get('commentCount', 0)
                data_hoje = datetime.now().strftime('%Y-%m-%d %H:%M')

            # Escrever essa linha do arquivo
                escritor.writerow([data_hoje, id_atual, titulo, views, likes, comments])
                print(f"Coletado com sucesso: {titulo[:20]}...") # Mostra só o começo do nome
            else:
                print(f"Vídeo {id_atual} não encontrado!")

        except Exception as e:
            print(f"Erro ao processar o vídeo {id_atual}: {e}")

print("--- FIM! Dados Adicionados ao histórico com sucesso. ---")
```
